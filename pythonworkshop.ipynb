```python
units = input("Enter the number of units consumed: ")
units = int(units)
customer_type = input("Enter customer type (residential, commercial, industrial): ").lower()

price = 0

if customer_type == "residential":
    price = units * 2
elif customer_type == "commercial":
    price = units * 5
elif customer_type == "industrial":
    price = units * 10
else:
    print("Invalid customer type entered.")
    price = 0 # Or handle as an error condition

if price > 0:
    print(f"For {units} units and a {customer_type} customer, the total energy price is: ${price:.2f}")
```
